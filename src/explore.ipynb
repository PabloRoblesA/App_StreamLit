```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

#Usar la suma de las caracteristicas ('total') para predecir la tasa de crecimiento ('growth_rate')
caracteristicas = df[['Total','HP','Attack','Defense','Sp. Atk','Sp. Def','Speed']]
pokeresultado = df['Legendary']

X_train, X_test, y_train, y_test = train_test_split(caracteristicas, pokeresultado, test_size=0.2, random_state=42)

scaler = StandardScaler()
X_train_scaler = scaler.fit_transform(X_train)
X_test_scaler = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

modelo_regrlog = LogisticRegression(max_iter=1000)

modelo_regrlog.fit(X_train_scaler, y_train)

y_pred = modelo_regrlog.predict(X_test_scaler)

accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
```
