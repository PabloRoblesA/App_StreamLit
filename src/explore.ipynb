```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

caracteristicas = df[['hp', 'attack',
       'defense', 'sp_atk', 'sp_def', 'speed', 'total','catch_rate']]
pokeresultado = df['growth_rate']

X_train, X_test, y_train, y_test = train_test_split(caracteristicas, pokeresultado, test_size=0.2, random_state=42)

scaler = StandardScaler()
scaler2 = MinMaxScaler()

X_train_scaler = scaler.fit_transform(X_train)
X_test_scaler = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

modelo_regrlog = LogisticRegression(max_iter=1000)

modelo_regrlog.fit(X_train_scaler, y_train)

y_pred = modelo_regrlog.predict(X_test_scaler)

accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
```

```python
from sklearn.tree import DecisionTreeClassifier

modelo_arbol = DecisionTreeClassifier(random_state=42)

modelo_arbol.fit(X_train, y_train)

y_pred = modelo_arbol.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
```

```python
from sklearn.ensemble import RandomForestClassifier

modelo_bosque = RandomForestClassifier(n_estimators=100, random_state=42)

modelo_bosque.fit(X_train_scaler, y_train)

y_pred = modelo_bosque.predict(X_test_scaler)

accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
#Seleccionado para mejorar hiperparametros
```

```python
from sklearn.ensemble import GradientBoostingClassifier

modelo_boosting = GradientBoostingClassifier(random_state=42)

modelo_boosting.fit(X_train_scaler, y_train)

y_pred = modelo_boosting.predict(X_test_scaler)

accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
```

```python
from sklearn.naive_bayes import GaussianNB

modelo_naive = GaussianNB()

modelo_naive.fit(X_train_scaler, y_train)

y_pred = modelo_naive.predict(X_test_scaler)

accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
```

```python
from sklearn.neighbors import KNeighborsClassifier

modelo_vecinos = KNeighborsClassifier(n_neighbors=5)

modelo_vecinos.fit(X_train_scaler, y_train)

y_pred = modelo_vecinos.predict(X_test_scaler)

accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
```
